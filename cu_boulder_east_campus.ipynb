```python
import osmnx as osm
import geopandas as gpd
import matplotlib.pyplot as plt
import geoviews
```

```python
# Look for everything with CU Boulder name
cu_boulder_gdf = osm.geocode_to_gdf("University of Colorado Boulder")
cu_boulder_gdf
```

```python
# Look for everything with CU Boulder name
cu_gdf = osm.features_from_address(
    'University of Colorado, CO, United States',
    {'amenity': ['university']},
    dist=50000)
cu_gdf
```

```python
# Find CU Boulder (East Campus) from OSM ()
east_campus = osm.geocode_to_gdf("University of Colorado Boulder (East Campus)")
east_campus
```

```python
# Plot the data
east_campus.plot()
```

```python
# Extract ploygon from geometry
polygon = east_campus.geometry.iloc[0]
polygon
```

```python
# Get buildings
buildings = osm.features_from_polygon(polygon,
                                      tags={"building": True})
buildings
```

```python
# Get roads
roads = osm.features_from_polygon(polygon,
                                      tags={"highway": True})
roads
```

```python
# Clip roads to campus boundary
roads_clipped = gpd.clip(
    roads,
    east_campus
)

roads_clipped
```

```python
# Keep only linestring for interactive plotting
road_lines = roads_clipped[
    roads_clipped.geometry.type.isin(["LineString", "MultiLineString"])]
road_lines
```

```python
# Get water
water = osm.features_from_polygon(polygon,
                                      tags={"natural":"water",
                                            "waterway": True,
                                            "water":True})
water
```

```python
# Clip water lines to campus boundary
water_clipped = gpd.clip(
    water,
    east_campus
)

water_clipped
```

```python
# Splitting water into lines and polygons for easier plotting
water_polygons = water_clipped[
    water_clipped.geometry.geom_type.isin(
        ["Polygon", "MultiPolygon"])]

water_polygons


water_lines = water_clipped[
    water_clipped.geometry.geom_type.isin(
        ["LineString", "MultiLineString"])]

water_lines
```

```python
# Plot
fig, ax = plt.subplots(figsize=(10, 10))

east_campus.plot(
    ax=ax,
    color="None",
    linewidth=2
)

buildings.plot(
    ax=ax,
    color="lightgray",
    edgecolor="black",
    linewidth=0.5
)

road_lines.plot(
    ax=ax,
    color="black",
    linewidth=0.5)

water_clipped.plot(
    ax=ax,
    color="blue",
    linewidth=0.5)

ax.set_title("Buildings, water, and roads of CU Boulder (East Campus) from OpenStreetMap")
ax.set_axis_off()

plt.show()
```

```python
import hvplot.pandas

buildings_map = buildings.hvplot(
    geo=True,
    tiles="OSM",
    fill_color="yellow",
    line_color="black",
    alpha=1.0,
    line_width=2,
    height=600,
    width=800,
    title="CU Boulder (East Campus)"
)

east_campus_map = east_campus.hvplot(
    geo=True,
    fill_color="yellow",
    line_color="black",
    fill_alpha=0.3,
    line_width=5
)

creek_map = water_lines.hvplot(
    geo=True,
    line_color="blue",
    line_width=4
)

pond_map = water_polygons.hvplot(
    geo=True,
    line_color="blue",
    fill_color="blue",
    fill_alpha=0.3,
    line_width=2
)

road_map = road_lines.hvplot(
    geo=True,
    line_color="black",
    line_width=1,
    alpha=0.8
)




east_campus_map = buildings_map * east_campus_map * creek_map * pond_map * road_map
east_campus_map
```
